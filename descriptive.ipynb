```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger les CSV
salah = pd.read_csv("statsheet_Mohamed_Salah.csv")
saka = pd.read_csv("statsheet_Bukayo_Saka.csv")
isak = pd.read_csv("statsheet_Alexander_Isak.csv")
mbeumo = pd.read_csv("statsheet_Bryan_Mbeumo.csv")
diaz = pd.read_csv("statsheet_Luis_DiÃÅaz.csv")
haaland = pd.read_csv("statsheet_Erling_Haaland.csv")
son = pd.read_csv("statsheet_Heung-min_Son.csv")
cunha = pd.read_csv("statsheet_Matheus_Cunha.csv")
johnson = pd.read_csv("statsheet_Brennan_Johnson.csv")
gordon = pd.read_csv("statsheet_Anthony_Gordon.csv")

plt.figure(figsize=(10, 6))

plt.plot(salah['Goals'], label='Mohamed Salah',color='red')
plt.plot(saka['Goals'], label='Bukayo Saka',color='red')
plt.plot(isak['Goals'], label='Alexander Isak',color='black')
plt.plot(mbeumo['Goals'], label='Bryan Mbeumo')
plt.plot(diaz['Goals'], label='Luis D√≠az',color='red')
plt.plot(haaland['Goals'], label='Erling Haaland',color='navy')
plt.plot(son['Goals'], label='Heung-min Son')
plt.plot(cunha['Goals'], label='Matheus Cunha')
plt.plot(johnson['Goals'], label='Brennan Johnson')
plt.plot(gordon['Goals'], label='Anthony Gordon')

# Ajouter des labels et un titre
plt.ylabel('Nombre de buts')
plt.xlabel('Matchs')
plt.title('√âvolution des buts de chaque joueur au fil des matchs')

# Ajouter une l√©gende
plt.legend()

# Afficher le graphique
plt.xticks(rotation=45)  # Rotation des dates pour une meilleure lisibilit√©
plt.tight_layout()  # Ajuster l'espace pour √©viter que le texte ne soit coup√©
plt.show()
```
