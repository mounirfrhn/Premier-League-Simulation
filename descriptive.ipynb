### Now that we've successfully added our statsheets, it would be nice to get a grasp on them -- visually.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
salah=pd.read_csv("statsheet_Mohamed_Salah.csv")
saka=pd.read_csv("statsheet_Bukayo_Saka.csv")
isak=pd.read_csv("statsheet_Alexander_Isak.csv")
mbeumo=pd.read_csv("statsheet_Bryan_Mbeumo.csv")
diaz=pd.read_csv("statsheet_Luis_DiÃÅaz.csv")
haaland=pd.read_csv("statsheet_Erling_Haaland.csv")
son=pd.read_csv("statsheet_Heung-min_Son.csv")
cunha=pd.read_csv("statsheet_Matheus_Cunha.csv")
johnson=pd.read_csv("statsheet_Brennan_Johnson.csv")
gordon=pd.read_csv("statsheet_Anthony_Gordon.csv")
```

```python
plt.figure(figsize=(10, 6))

plt.plot(salah['Goals'] + salah['Assists'], label='Mohamed Salah',color='red')
plt.plot(saka['Goals']+saka['Assists'], label='Bukayo Saka',color='green')
plt.plot(isak['Goals']+isak['Assists'], label='Alexander Isak',color='black')
plt.plot(mbeumo['Goals']+mbeumo['Assists'], label='Bryan Mbeumo', color='pink')
plt.plot(diaz['Goals']+diaz['Assists'], label='Luis D√≠az',color='red')
plt.plot(haaland['Goals']+haaland['Assists'], label='Erling Haaland',color='skyblue')
plt.plot(son['Goals']+son['Assists'], label='Heung-min Son', color='navy')
plt.plot(cunha['Goals']+cunha['Assists'], label='Matheus Cunha', color='yellow')
plt.plot(johnson['Goals']+johnson['Assists'], label='Brennan Johnson', color="navy")
plt.plot(gordon['Goals']+gordon['Assists'], label='Anthony Gordon', color='black')

plt.ylabel('Nombre de buts+ P/D')
plt.xlabel('Matchs')
plt.title('√âvolution des buts + P/D de chaque joueur au fil des matchs')

plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Calculer la diff√©rence entre Goals et xG pour chaque joueur
salah_diff = salah['Goals'] - salah['xG']
saka_diff = saka['Goals'] - saka['xG']
isak_diff = isak['Goals'] - isak['xG']
mbeumo_diff = mbeumo['Goals'] - mbeumo['xG']
diaz_diff = diaz['Goals'] - diaz['xG']
haaland_diff = haaland['Goals'] - haaland['xG']
son_diff = son['Goals'] - son['xG']
cunha_diff = cunha['Goals'] - cunha['xG']
johnson_diff = johnson['Goals'] - johnson['xG']
gordon_diff = gordon['Goals'] - gordon['xG']

plt.figure(figsize=(10, 6))

plt.plot(salah_diff, label='Mohamed Salah', color='red')
plt.plot(saka_diff, label='Bukayo Saka', color='green')
plt.plot(isak_diff, label='Alexander Isak', color='black')
plt.plot(mbeumo_diff, label='Bryan Mbeumo', color='pink')
plt.plot(diaz_diff, label='Luis D√≠az', color='red')
plt.plot(haaland_diff, label='Erling Haaland', color='skyblue')
plt.plot(son_diff, label='Heung-min Son', color='navy')
plt.plot(cunha_diff, label='Matheus Cunha', color='yellow')
plt.plot(johnson_diff, label='Brennan Johnson', color="navy")
plt.plot(gordon_diff, label='Anthony Gordon', color='black')

# Ajouter des √©tiquettes et un titre
plt.ylabel('Diff√©rence (Goals - xG)')
plt.xlabel('Matchs')
plt.title('Diff√©rence entre Goals et xG de chaque joueur')

plt.legend()
plt.tight_layout() 
plt.show()
```
